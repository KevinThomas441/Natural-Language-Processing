Helper functions to read and preprocess data

```python
#Function to read files
def read_files(path):
    f = open(path, "r", encoding="utf-8")
    text = f.read()
    f.close()
    return text
```

```python
def preprocess(text):
    
    #Creating a variable to hold all the unwanted punctuations
    punctuations = '!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~'
    
    #Splitting the text into lines
    text = text.split("\n")
    
    #Removing punctuations
    text = [s.translate(str.maketrans('', '', punctuations)) for s in text]
    
    #Lowercasing the text
    text = [sentence.lower() for sentence in text]
    
    return text
```

A dictionary or dictionaries is used to store the translational probabilities. The inner dictionary stores all of the French words while the outer dictionary stores the English words. This data structure is used because it helps translate from mathematical terminology and pseudocode easily.<br>
For example:<br>
t(e|f) = t ['english']['french']<br>

We need to initialize all translational probabilities uniformly. This is done by assigning 1/(length of unique target words) to each word pair.

```python
#Function to initialize translation probabilities uniformly
def initialize_probabilities(source_words, target_words):
    
    #Setting the transitional probabilities to 1/(length of unique target words)
    t = {target_word: {source_word: 1/len(target_words) for source_word in source_words} for target_word in target_words}
    
    #Return probabilities
    return t
```

```python
t = initialize_probabilities(english_words, french_words)
```

To display probability tables, we convert our probability dictionary into a pandas DataFrame. This is done due to Pandas' built in tabular format which enables easy visualization.

```python
#Function to print the probability table between two coorpora
def print_probabilities(t):
    
    #Transforming the dictionary into a Pandas dataframe
    table = pd.DataFrame.from_dict(t)
    
    #Return Table
    return table
```

```python
print_probabilities(t)
```

Pseudo-Code:<br>

 initialize t(e|f) uniformly<br>
 do until convergence<br>
   set count(e|f) to 0 for all e,f<br>
   set total(f) to 0 for all f<br>
   for all sentence pairs (e_s,f_s)<br>
     set total_s(e) = 0 for all e<br>
     for all words e in e_s<br>
       for all words f in f_s<br>
         total_s(e) += t(e|f)<br>
     for all words e in e_s<br>
       for all words f in f_s<br>
         count(e|f) += t(e|f) / total_s(e)<br>
         total(f)   += t(e|f) / total_s(e)<br>
   for all f<br>
     for all e<br>
       t(e|f) = count(e|f) / total(f)<br>

Eucidean Distance is used as a metric to check for convergence. If the distance between the previous probabilities and the current probabilities is less than a chosen threshold, it can be said that convergence has been achieved.

```python
#Function to calculate Euclidean Distance
#Reference: https://www.techtrekking.com/how-to-calculate-euclidean-and-manhattan-distance-by-using-python/
def euclidean_distance(prev_t, t):
    
    #Euclidean Distance = ((y1-y1Hat)**2 + (y2-y2Hat)**2 .... (yn-ynHat)**2)**0.5
    
    #Getting the source keys
    e = t.keys()
    
    #Getting the target keys
    f_temp = list(t.values())
    f = f_temp[0].keys()
    
    result = 0
    
    #Iterating over all words in source and 
    for (e, f) in zip(e, f):
        
        #Adding up the squares of differences
        delta = (prev_t[e][f] - t[e][f]) ** 2
        result += delta
    
    #Returning the square root of the sum of squares of differences 
    return result ** 0.5
```

```python
def calculate_probabilities(source, target, source_words, target_words):
    
    print("Initializing")
    t = initialize_probabilities(source_words, target_words)
    
    #Creating a copy of initial probabilities
    prev_t = copy.deepcopy(t)
    converged = False
    iterations = 0
    
    print("Starting")
    #Repeat until probabilitiees stop changing
    while not converged:
        iterations+=1

        #Setting count = 0 for all (Source|Target) pairs
        count = {target_word: {source_word: 0 for source_word in source_words} for target_word in target_words}

        #Setting total = 0 for all Source words
        total = {source_word: 0 for source_word in (source_words)}

        #Iterating over all sentence pairs
        for i in range(len(source)):

            #Iterating over all words in the Target sentence
            for target_word in target[i].split():
                total_sent = 0

                #Iterating over all words in Source sentence
                for source_word in source[i].split():

                    #Counting the total number of alignments possible
                    total_sent += t[target_word][source_word]

                #Iterating over all words in Source sentence
                for source_word in source[i].split():

                    #Updating the count for each word pair
                    count[target_word][source_word] += t[target_word][source_word] / total_sent

                    #Updating the total number of alignments possible for each Source word
                    total[source_word] += t[target_word][source_word] / total_sent

        #Iterating over all unique Source words
        for source_word in source_words:

            #Iterating over all unique Traget words
            for target_word in target_words:

                #Updating the translational probabilities of each (Source|Target) word pair
                t[target_word][source_word] = count[target_word][source_word] / total[source_word]

        #Checking for convergence
        if euclidean_distance(prev_t, t)<0.002:
            converged = True

        #Creating a copy of translational probabilities to check for convergence in next iteration
        prev_t = copy.deepcopy(t)

        #Printing the iteration number
        print("Iteration: ", iterations)
    print("Done!")    
    return t
```

```python
#Calculating translation probabilities for French English
t = calculate_probabilities(english, french, english_words, french_words)
```

```python
#Printing out most probable words
#Reference: https://stackoverflow.com/questions/268272/getting-key-with-maximum-value-in-dictionary
import operator
print("{0:20}{1:20}{2:20}".format('French', 'English', 'Probability'))
print("-------------------------------------------------------------")
for french_word in (french_words):
    print("{0:20}{1:20}{2:20}".format(french_word, max(t[french_word].items(), key=operator.itemgetter(1))[0], max(t[french_word].items(), key=operator.itemgetter(1))[1]))
```
