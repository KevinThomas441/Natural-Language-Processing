### Step 1:
Importing relevant libraries and the first 300 words from Alice in Wonderland

```python
import nltk
from nltk.corpus import gutenberg
from collections import Counter
corp = gutenberg.words("carroll-alice.txt")
corp = corp[0:300]
```

### Step 2:
Using the POS tagger from the nltk library

```python
corp_tagged = nltk.pos_tag(corp)
corp_tagged
```

### Step 3:
Using Counter from collections to order the parts of speech based on frequency, then using the most_common() method of the Counter to show the five most common ones.

```python
pos_freq = Counter([x[1] for x in corp_tagged])
```

```python
pos_freq.most_common(5)
```

The most common parts of speech are: IN (Preposition), NN (Singular Noun), DT (Determinant), NNP (Singular Proper Noun) and PRP (Personal Pronouns)

### Step 4:
Using RegexpTagger from the nltk library to tag parts of speeck based on the patterns inputted by us.

```python
patterns = [(r'.*ing$', 'VBG'), (r'.*ed$', 'VBD'), (r'.*s$', 'NNS')]
```

```python
regexp_tagger = nltk.RegexpTagger(patterns)
```

```python
regexp_tagger.tag(corp)
```

<ul>
    <li>The regexp tagger tags the word "was" as NNS (plural noun) whereas the pos tagger tags it as VBD (simple past)</li>
    <li>The regexp tagger tags the word "as" as NNS (plural noun) whereas the pos tagger tags it as RB (adverb)</li>
    <li>The regexp tagger tags the word "Lewis" as NNS (plural noun) whereas the pos tagger tags it as NNP (Singular Proper Noun)</li>
</ul>
